```python
import pandas as pd
import numpy as np
```

```python
## problem 1: data exploratory
df_receipts = pd.read_json("Fetch/receipts.json", lines=True)
df_receipts._id = pd.json_normalize(df_receipts._id)

df_receipts_items = pd.concat([df_receipts._id, df_receipts.rewardsReceiptItemList.explode().apply(pd.Series)], axis=1)

df_receipts_items = df_receipts_items[["_id", "brandCode", "rewardsProductPartnerId",
                       "deleted", "finalPrice", "partnerItemId", "pointsEarned",
                       "pointsNotAwardedReason", "preventTargetGapPoints", "quantityPurchased",
                       "rewardsGroup", "targetPrice"]].rename(columns={"_id": "receipt_id"})
df_receipts_items
```

```python
df_brand = pd.read_json("Fetch/brands.json", lines=True)
df_brand = pd.concat([df_brand, pd.json_normalize(df_brand.cpg)], axis=1)
df_brand._id = pd.json_normalize((df_brand._id))

df_brand = df_brand[["_id", "barcode", "category", "categoryCode", "name", "topBrand", "brandCode",
                     "$ref", "$id.$oid"]].rename(columns={"$ref": "cpg_ref", "$id.$oid": "cpg_id"})
df_brand
```

```python
df_brand2 = df_brand.dropna(subset=["brandCode","cpg_id"])
df_brand2 = df_brand2.loc[df_brand.brandCode != ""]
df_brand2
```

```python
df_brand2["combined"] = df_brand2["brandCode"] + df_brand2["cpg_id"]
df_brand2[df_brand2.duplicated(['combined'], keep=False)] ## proves that brandCode and cpg_id pair are composite key
```

```python
# problem 3: looking for data quality issue
# there are a lot of null values in brandCode field in both brand schema and receiptItemList schema

# 20% of df_brand's brandCode column are null values
print("null value percentage in brand schema's brandCode: \n")
print(str(df_brand[df_brand.brandCode.isnull()].shape[0]/df_brand.shape[0]*100) + "%")

# 64% of df_receipts_items's brandCode column are null values
print("null value percentage in receipt item list schema's brandCode: \n")
print(str(df_receipts_items[df_receipts_items.brandCode.isnull()].shape[0]/df_receipts_items.shape[0]*100) + "%")

# while my assumption of using brandCode and cpg_id/rewardsProductPartnerId to map df_brand to df_receipt_items seems to
# be correct, there are some brandCodes spelled differently in df_brand and df_receipt_items

print(sorted(pd.unique(df_brand.brandCode.dropna().astype(str)).tolist()))
print(sorted(pd.unique(df_receipts_items.brandCode.dropna().astype(str)).tolist()))
# from the above printed list of unique items in brandCode in both dataframes, we can see that, for example, Ben and
# Jerry's is spelled as Ben & Jerry's in df_brand, while in df_receipt_items, it's spelled as Ben and Jerrys
# This can be easily observed by seeing how many brands in df_receipts_items are not in df_brand by the following:
sorted_difference = sorted(list(set(pd.unique(df_receipts_items.brandCode.dropna().astype(str))) - \
                         set(pd.unique(df_brand.brandCode.dropna().astype(str)).tolist())))
print(sorted_difference)

# also note that around 80% of the brands in df_receipt_items are not included in df_brand, which is supposed to be the
# schema that contains all brands as shown below:
print("percentage of brandCodess in receipt item list schema that are not in the brand schema: \n")
print(str(len(sorted_difference)/len(sorted(pd.unique(df_receipts_items.brandCode.dropna().astype(str)).tolist()))*100)+"%")

# Inconsistency in spelling in the column brandCode can lead to implausible join between both dataframes

#
```
